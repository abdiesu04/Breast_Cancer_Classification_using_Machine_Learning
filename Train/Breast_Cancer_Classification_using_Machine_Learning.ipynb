```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
# loading rhe data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
#printing the first five row
data_frame.head()
```

```python
#adding the 'taget' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
# print the first five rows of the data frame
data_frame.head()
```

```python
#number of rows and clumns in the dataset
data_frame.shape
```

```python
#informations about the data
data_frame.info()
```

```python
#checking there no missing values
data_frame.isnull().sum()
```

```python
#measures about the data set
data_frame.describe()
```

```python
# counting the distribution of target variable
data_frame['label'].value_counts()
```

1 --- Represents Benign
2 --- Represents Malignant

```python

data_frame.groupby('label').mean()
# Note the Benign attributes are always greater than the Malignant attributes
```

Separating the features and target

```python
x = data_frame.drop(columns = 'label', axis = 1)
y = data_frame['label']
print(x , y)
```

**Visualizing the Result**

Splitting the data into training data and testing data

```python
import seaborn as sns

# Combine x and y to form a single DataFrame for visualization
data_frame['label'] = y
sns.pairplot(data_frame, hue='label', vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area'])
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)
#printing the size
print(x.shape, x_train.shape, x_test.shape)
print(y.shape, y_train.shape , y_test.shape)
```

**Model Traininig**
Logistic Regression

```python
model = LogisticRegression()
```

```python
#traning the Logistic regression model using traing the training data
model.fit(x_train, y_train)
```

**Model Evaluation**
Accuracy score

```python
#accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(y_train, x_train_prediction)
print('Accuracy on training data : ', training_data_accuracy)
```

```python
#accuracy on test data
x_test_prediction = model.predict(x_test)
training_data_accuracy = accuracy_score(y_test, x_test_prediction)
print('Accuracy on test data : ', training_data_accuracy)
```
